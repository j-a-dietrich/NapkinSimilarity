```python
import NapkinSimilarity
from matchms.importing import load_from_mgf

from itertools import product
from tqdm import tqdm
```

```python
test_spec = list(load_from_mgf("pesticides.mgf"))
```

```python
pairs = list(product(range(len(test_spec)), repeat=2))
```

```python
similarity = []
for idx1, idx2 in tqdm(pairs):
    sim = NapkinSimilarity.run(test_spec[idx1], test_spec[idx2])
    similarity.append(sim)
```
